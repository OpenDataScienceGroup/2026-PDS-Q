```python
import pandas as pd
import os
import glob
from functools import reduce

files = glob.glob("*.csv")

print(files)

dfs = []

for file in files:
    df = pd.read_csv(file)

    # Keep only needed columns
    df = df[["image_name", "hair", "pen_marks"]]

    # Get file name without extension
    name = os.path.splitext(os.path.basename(file))[0]

    # Rename columns except image_name
    df = df.rename(columns={
        "hair": f"hair_{name}",
        "pen_marks": f"pen_marks_{name}"
    })

    dfs.append(df)

# Merge all files on image_name
merged_df = reduce(
    lambda left, right: pd.merge(left, right, on="image_name", how="outer"),
    dfs
)

merged_df.to_csv("combined.csv", index=False)
```
