## Metadata Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("metadata.csv")
```

### 1. Age distribution

```python
#summary statistics for age

age_stats = df["age"].describe()
print("Age Summary Statistics:")
print(age_stats)
```

Patients ages vary from 17 to 84 years, with a median age of 64, with the majority falling between 52 and 72 years old.

```python
plt.hist(df["age"].dropna(), bins=10, color='skyblue', edgecolor='black')
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Number of patients")
plt.show()
```

from the histogram, the majority of the patients are between 60 and 80 years old, with a peak around 65-75, and very few are younger than 40. As a result, the dataset is skewed towards older people.

### 2. Gender Distribution

```python
print(df["gender"].value_counts())
```

```python
df["gender"].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightpink'])
plt.ylabel("")
plt.title("Gender Distribution")
plt.show()
```

The dataset has more male patients(57.5%) than female patients(42.5%). This imbalance could be caused by sampling, disease prevalence, or random variation in the population.

### 3. Diagnosis Distribution

```python
print(df["diagnostic"].value_counts())
```

```python
df["diagnostic"].value_counts().plot.bar(color= 'lightgreen', edgecolor= 'black')
plt.title("Diagnosis distribution")
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.show()
```

The dataset`s most common diagnosis is **BCC ,** which accounts for around 45 cases. The second most common is **ACK**, which accounts for approximately 35 instances. Other diagnosis  **NEV**, **SCC**, and **SEK** occur significantly less frequent, with approximately 10-12 cases each. **MEL** is the least represented, appearing less than five times. Overall, the dataset consists primarily of elderly male patients diagnosed with BCC or ACK, which is consistent with diseases associated with prolonged sun exposure over time.
